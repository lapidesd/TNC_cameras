```python
import pandas as pd
import numpy as np
```

```python
daymet = pd.read_csv("Daymet2006_2024.csv")
daymet["date"] = pd.to_datetime(daymet["year"] * 1000 + daymet["yday"], format="%Y%j")
```

```python
daymet["ea_kPa"] = daymet["vp"] / 1000

# Compute saturation vapor pressure (Tetens equation)
def es(T):
    return 0.6108 * np.exp((17.27 * T) / (T + 237.3))

daymet["es_tmax"] = es(daymet["tmax"])
daymet["es_tmin"] = es(daymet["tmin"])
daymet["es_mean"] = (daymet["es_tmax"] + daymet["es_tmin"]) / 2
daymet["tvpd"] = daymet["es_mean"] - daymet["ea_kPa"]

daymet = daymet.drop(columns = ["ea_kPa", "es_tmax", "es_tmin", "es_mean"])
```
